### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
query_url = f'{url}api-key={nyt_api_key}&begin_date={begin_date}&end_date={end_date}&fq={filter_query}&sort={sort}&fl={field_list}'
query_url
```

```python
#query_url = f'{query_url}&page=0'

#test_response = requests.get(query_url)
#test_review = test_response.json()
#test_review
```

```python
# Create an empty list to store the reviews

reviews_list = []

# loop through pages 0-19
for page in range(0,20):                                            # taking a loop that will go inside our API
                                                                    #    

    
    page_query_url = f'{query_url}&page={page}'                      # specifiys which page we are looking for
    response = requests.get(page_query_url)
    reviews = response.json()

    # create query with a page number
    # API results show 10 articles at a time
    # Make a "GET" request and retrieve the JSON
    

    
    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)
    
    # Try and save the reviews to the reviews_list
    try:                                                               # trying to check that those pages we are looping actually exist
                                                                        # the try is going to look inside the pages
                                                                        # its looking in the range of 1 to 19
                                                                        

        for result in reviews['response']['docs']:
            reviews_list.append(result)        
        # loop through the reviews["response"]["docs"] and append each review to the list
    
        # Print the page that was just retrieved
        print(f'This is your result {page}')

        # Print the page number that had no results then break from the loop
    except:
            print(f'No result found. {page}')

            
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
# slice through results

print(json.dumps(reviews_list[0:5] ,indent=4))
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
converted_list_df =  pd.json_normalize(reviews_list)
converted_list_df.head()
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early
converted_list_df['title'] = converted_list_df['headline.main'].apply(lambda st: st[st.find('\u2018')+1:st.find('\u2019 Review')])
converted_list_df.head()
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
converted_list_df['keywords']=converted_list_df['keywords'].apply(extract_keywords)
converted_list_df.head()
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database
titles = converted_list_df['title'].to_list()
titles
```

### Access The Movie Database API

```python
#converted_list_df.columns
```

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
details_url = "https://api.themoviedb.org/3/movie/"
tmdb_key_string = "&api_key=" + tmdb_api_key
```

```python
#test_movie_title = 'The Attachment Diaries'
#test_query_url = f'{url}{test_movie_title}{tmdb_key_string}'
#test_reponse = requests.get(test_query_url)
#test_movie_info = test_reponse.json()
#print(json.dumps(test_movie_info, indent=4))
```

```python
# Create an empty list to store the results
tmdb_movies_list = []

# Create a request counter to sleep the requests after a multiple
# of 50 requests
request_counter = 0


# Loop through the titles
for title in titles:
   #test break 
   #if request_counter == 6:
    #break
    # Check if we need to sleep before making a request
    if request_counter % 50 == 0:
        print(f'Requests is sleeping.')
        time.sleep(1)
    # creating a time to access our code at. 
    # checking if we need to sleep before making our api request. 
    # Add 1 to the request counter
    request_counter+=1
    
    # Perform a "GET" request for The Movie Database
    
    #title_query_url = f'{url}{title}{tmdb_key_string}'
    title_query_url = url+title+tmdb_key_string
    #print(title_query_url)
    title_response = requests.get(title_query_url)
    movie = title_response.json()


    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie
    # is not found.
    try:

        # Get movie id
        #print(movie)
        movie_id = movie['results'][0]['id'] 




        # Make a request for a the full movie details
        detail_query_url = f'{details_url}{movie_id}?{tmdb_key_string}'
        #print(details_url)
        #print(movie_id)
        #print(tmdb_key_string)
        #print(detail_query_url)

        # Execute "GET" request with url

        detail_response = requests.get(detail_query_url)
        #detail_response.status_code
        movie_details = detail_response.json()

        # Extract the genre names into a list

        genres_list = []
        for index in range(len(movie_details['genres'])):
            for key in movie_details['genres'][index]:
                    if key == 'name':
                        genres_list.append(movie_details['genres'][index][key])

        # Extract the spoken_languages' English name into a list
        spoken_language_list = []
        for index in range(len(movie_details['spoken_languages'])):
                for key in movie_details['spoken_languages'][index]:
                    if key == 'english_name':
                        spoken_language_list.append(movie_details['spoken_languages'][index][key])

        # Extract the production_countries' name into a list
        production_countries_list = []
        for index in range(len(movie_details['production_countries'])):
            for key in movie_details['production_countries'][index]:
                if key == 'name':
                    production_countries_list.append(movie_details['production_countries'][index][key])

        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list
        tmdb_movies_list.append({'title': title,
                                'original_title': movie_details['original_title'],
                                'budget': movie_details['budget'],
                                'original_language': movie_details['original_language'],
                                'homepage': movie_details['homepage'],
                                'overview': movie_details['overview'],
                                'popularity': movie_details['popularity'],
                                'runtime': movie_details['runtime'],
                                'revenue': movie_details['revenue'],
                                'release_date': movie_details['release_date'],
                                'vote_average': movie_details['vote_average'],
                                'vote_count': movie_details['vote_count'],
                                'genres': genres_list,
                                'spoken_languages': spoken_language_list,
                                'production_countries': production_countries_list
                            })

        # Print out the title that was found
        print(f'The Movie Title: "{title}" was found.')

    except:
        print(f"The Movie Title: '{title}' was not found.")
        
    
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
print(json.dumps(tmdb_movies_list[0:5], indent=4))
```

```python
# Convert the results to a DataFrame
tmdb_df = pd.DataFrame(tmdb_movies_list)
tmdb_df.head()
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title

merged_nyt_tmdb_df = pd.merge(converted_list_df, tmdb_df, on='title', how='inner')
merged_nyt_tmdb_df.head()
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
columns_to_fix = ['genres', 'spoken_languages', 'production_countries']


# Create a list of characters to remove
character_to_remove = ['[',']', '"']

# Loop through the list of columns to fix
for col in columns_to_fix: 
    # Convert the column to type 'str'
    merged_nyt_tmdb_df[col] = merged_nyt_tmdb_df[col].astype('str')

    # Loop through characters to remove
for char in character_to_remove:
    merged_nyt_tmdb_df[col] = merged_nyt_tmdb_df[col].str.replace(char, '', regex=False)
    

# Display the fixed DataFrame
    merged_nyt_tmdb_df.head()
```

```python
# Drop "byline.person" column
merged_nyt_tmdb_df = merged_nyt_tmdb_df.drop(['byline.person'], axis = 1)
merged_nyt_tmdb_df.head()
```

```python
# Delete duplicate rows and reset index
merged_nyt_tmdb_df = merged_nyt_tmdb_df.drop_duplicates()
merged_nyt_tmdb_df = merged_nyt_tmdb_df.reset_index(drop=True)
merged_nyt_tmdb_df.head()
```

```python
# Export data to CSV without the index
merged_nyt_tmdb_df.to_csv('merged_nyt_tmdb_df.csv', encoding='utf-8', index=False)
```
